```python
from tkinter import *
calculation = "" 
def press(num):  
    global calculation
    calculation = calculation + str(num) 
    equation.set(calculation) 
def equalpress(): 

    try: 
 
        global calculation 
 

        total = str(eval(calculation)) 
 
        equation.set(total) 

        calculation = "" 

    except: 
 
        equation.set(" error ") 
        calculation = "" 
def clear(): 
    global calculation 
    calculation = "" 
    equation.set("") 
 
 
# Driver code 
if __name__ == "__main__": 
    tk = Tk() 
    tk.configure(background="black") 
    tk.title("Simple Calculator") 
    tk.geometry("255x300") 
    equation = StringVar()  
    equation_field = Entry(tk, textvariable=equation) 
    equation_field.grid(columnspan=10, ipadx=70, ipady=15)
    button1 = Button(tk, text=' 1 ', fg='black', bg='white', 
                    command=lambda: press(1), height=2, width=7,borderwidth = 5) 
    button1.grid(row=2, column=0) 
 
    button2 = Button(tk, text=' 2 ', fg='black', bg='white', 
                    command=lambda: press(2), height=2, width=7,borderwidth = 5) 
    button2.grid(row=2, column=1) 
 
    button3 = Button(tk, text=' 3 ', fg='black', bg='white', 
                    command=lambda: press(3), height=2, width=7,borderwidth = 5) 
    button3.grid(row=2, column=2) 
 
    button4 = Button(tk, text=' 4 ', fg='black', bg='white', 
                    command=lambda: press(4), height=2, width=7,borderwidth = 5) 
    button4.grid(row=3, column=0) 
 
    button5 = Button(tk, text=' 5 ', fg='black', bg='white', 
                    command=lambda: press(5), height=2, width=7,borderwidth = 5) 
    button5.grid(row=3, column=1) 
 
    button6 = Button(tk, text=' 6 ', fg='black', bg='white', 
                    command=lambda: press(6), height=2, width=7,borderwidth = 5) 
    button6.grid(row=3, column=2) 
 
    button7 = Button(tk, text=' 7 ', fg='black', bg='white', 
                    command=lambda: press(7), height=2, width=7,borderwidth = 5) 
    button7.grid(row=4, column=0) 
 
    button8 = Button(tk, text=' 8 ', fg='black', bg='white', 
                    command=lambda: press(8), height=2, width=7,borderwidth = 5) 
    button8.grid(row=4, column=1) 
 
    button9 = Button(tk, text=' 9 ', fg='black', bg='white', 
                    command=lambda: press(9), height=2, width=7,borderwidth = 5) 
    button9.grid(row=4, column=2) 
 
    button0 = Button(tk, text=' 0 ', fg='black', bg='white', 
                    command=lambda: press(0), height=2, width=7,borderwidth = 5) 
    button0.grid(row=5, column=0) 
 
    plus = Button(tk, text=' + ', fg='black', bg='white', 
                command=lambda: press("+"), height=2, width=7,borderwidth = 5) 
    plus.grid(row=2, column=3) 
 
    minus = Button(tk, text=' - ', fg='black', bg='white', 
                command=lambda: press("-"), height=2, width=7,borderwidth = 5) 
    minus.grid(row=3, column=3) 
 
    multiply = Button(tk, text=' * ', fg='black', bg='white', 
                    command=lambda: press("*"), height=2, width=7,borderwidth = 5) 
    multiply.grid(row=4, column=3) 
 
    divide = Button(tk, text=' / ', fg='black', bg='white', 
                    command=lambda: press("/"), height=2, width=7,borderwidth = 5) 
    divide.grid(row=5, column=3) 
 
    equal = Button(tk, text=' = ', fg='black', bg='white', 
                command=equalpress, height=2, width=7,borderwidth = 5) 
    equal.grid(row=5, column=2) 
 
    clear = Button(tk, text='Clear', fg='black', bg='white', 
                command=clear, height=2, width=7,borderwidth = 5) 
    clear.grid(row=6, column=0) 
 
    Decimal= Button(tk, text='.', fg='black', bg='white', 
                    command=lambda: press('.'), height=2, width=7,borderwidth = 5) 
    Decimal.grid(row=5, column=1)  
    tk.mainloop() 
```
